```python
import asyncio
import random
import logging
import time
from concurrent.futures import ThreadPoolExecutor

logging.basicConfig(level="INFO")


def fct(name):
    value = random.randint(1, 7)
    logging.info(f"Sleeping {value} seconds for coroutine {name}")
    time.sleep(value)
    logging.info(f"Return value for coroutine {name}")

    return name + "_finished"


async def main():
    loop = asyncio.get_event_loop()
    with ThreadPoolExecutor(max_workers=40) as executor:
        methods = [loop.run_in_executor(executor, fct, f"coroutine_{el}") for el in range(1, 100)]
        methods_gathered = asyncio.gather(*methods)

    return await methods_gathered


loop = asyncio.get_event_loop()
res = loop.create_task(main())
```

```python
fct_results = res.result()
```
